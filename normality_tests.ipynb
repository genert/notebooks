```python
import pandas as pd
import numpy as np
import pandas_datareader.data as web

symbols = ["^GDAXI", "^GSPC", "MSFT", "PYPL"]
```

```python
data = pd.DataFrame()
for symbol in symbols:
    data[symbol] = web.DataReader(symbol, data_source="yahoo",start="1/1/2000")["Adj Close"]
```

```python
(data / data.iloc[0] * 100).plot(figsize=(8,6))
```

```python
log_returns = np.log(data / data.shift(1))
log_returns.hist(bins=50, figsize=(9, 6))
```
